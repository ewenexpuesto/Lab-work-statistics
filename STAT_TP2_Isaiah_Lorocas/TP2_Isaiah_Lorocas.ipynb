```r
x <- pnorm(35,20,sqrt(60),FALSE)
x
```

```r
# Paramètres
n <- 20
mu <- 20
sigma <- sqrt(60)

# Simulation de l'échantillon
sample_x <- rnorm(n, mean = mu, sd = sigma)

# Probabilité empirique que X > 35
empirical_prob <- sum(sample_x > 35) / n
empirical_prob

# Histogramme
hist(sample_x, probability = TRUE, col = "lightblue", breaks = 10, 
     main = "Histogramme de l'échantillon avec densité théorique",
     xlab = "Valeurs de X")

# Superposition de la densité théorique
curve(dnorm(x, mean = mu, sd = sigma), add = TRUE, col = "red", lwd = 2)
```

```r
n_sim <- 10
datmat <- matrix(0, nrow = n, ncol = n_sim)
empirical_probs <- numeric(n_sim)
for (i in 1:n_sim) {
  datmat[, i] <- rnorm(n, mean = mu, sd = sigma)  # Simulation
  empirical_probs[i] <- sum(datmat[, i] > 35) / n  # Proba empirique
}
empirical_probs


par(mfrow = c(2, 5))  # Disposition des graphiques (2 lignes, 5 colonnes)
for (i in 1:n_sim) {
  hist(datmat[, i], probability = TRUE, col = "lightblue", breaks = 10,
       xlim = c(mu - 3 * sigma, mu + 3 * sigma), ylim = c(0, 0.07),
       main = paste("Échantillon", i), xlab = "Valeurs de X")
  curve(dnorm(x, mean = mu, sd = sigma), add = TRUE, col = "red", lwd = 2)
}
```
