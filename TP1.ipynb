```r

x<-seq(3, 20,by=4)
sum_x<-0
n<-length(x)
for(i in 1:n){
  sum_x=sum_x+x[i]^2
}
sum_x

my_smean <- function(x)
{
  sum_x <- 0
  n <- length(x)
  for (i in 1:n){
    sum_x <- sum_x + x[i]
  }
  return(sum_x/n)
}
my_smean(x)

my_stdev <- function(x)
{
  n <- length(x)
  mean_x <- my_smean(x)
  sigma_squared <- 0
  for (i in 1:n){
    sigma_squared <- sigma_squared + (x[i] - mean_x)^2
  }
  sigma_squared <- sigma_squared / n
  return(sigma_squared^(1/2))
}
stdev_x <- my_stdev(x)
stdev_x

x <- rnorm(30,0,0.5)
y <- runif(25,-2,2)
x
y

plot(x)
barplot(x)
boxplot(y)
hist(y)

```
